```python
import numpy as np
import matplotlib.pyplot as plt
import csv

address = r'C:\Users\joela\Desktop\LanziPython/test.csv' #csv file address
start_time = 0 ## start of array 
end_time = 0 ## end of array 
on_off=[] ## array for on/off neuromorphic camera values
timestamp=[] ## array for timestamp value of camera samples

with open(address) as csvfile:
    spamreader = csv.reader(csvfile, delimiter=',')
    for row in spamreader:
        on_off.append(row[0])
        timestamp.append(row[1])
        try:
            on_off[-1]=int(on_off[-1])
            timestamp[-1]=int(timestamp[-1])
        except:
            continue

## grab header values from csv
eventTitle=on_off.pop(0)
timeStampTitle=timestamp.pop(0)

## remove stray startup value.. maybe
## del on_off[0:2]
## del timestamp[0:2]

## normalize time stamps
first_time=timestamp[0]
for x in range(len(timestamp)):
    timestamp[x]=timestamp[x]-first_time
```
