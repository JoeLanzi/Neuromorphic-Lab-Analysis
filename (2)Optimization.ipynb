```python
import numpy as np
import pandas as pd

#Importing CVS into DataFrame
data = pd.read_csv('test.csv')

#Converting DataFrame columns into list
on_off = list(data['On/Off']) #list used instead of series
timestamp = list(data['Timestamp']) #list used instead of series
```

```python
#Sample of DataFrame
data.loc[0:5]
```

```python
#Sample of List
on_off[0:10]
```

```python
## normalizing time stamps
first_time=timestamp[0]
for x in range(len(timestamp)):
    timestamp[x]=timestamp[x]-first_time
```

```python
#Sample of normalized Timestamp
timestamp[:5]
```

```python
## extracting one period from data
one_period = [i for i in timestamp if i < 28500]    
end_time = len(one_period)
start_time =0

len(one_period) #length of one period
```

```python
#Graph of CSV File
import matplotlib.pyplot as plt
%matplotlib notebook

plt.bar(timestamp[start_time:end_time],on_off[start_time:end_time],width=90)
plt.show()
```
