```python
import pandas as pd
import numpy as np
from plotly import __version__
import cufflinks as cf
from plotly.offline import download_plotlyjs,init_notebook_mode,plot,iplot
```

```python
init_notebook_mode(connected=True)
cf.go_offline()

%matplotlib inline
```

```python
data1 = pd.read_csv('test1.csv')
```

```python
data1.head()
```

```python
on_off = list(data1['On/Off'])
timestamp = list(data1[' Timestamp'])
x_axis = list(data1[' X'])
y_axis = list(data1[' Y'])
```

```python
# Normalizes timestamp
first_time1=timestamp[0]
for x in range(len(timestamp)):
    timestamp[x]=timestamp[x]-first_time1
```

```python
# Normalizes timestamp in dataframe
data1[' Timestamp'] = timestamp
```

```python
data1.head()
```

```python
#Calculates coordinate location
polarity_number = []
for x in range(len(timestamp)):
    polarity_number.append((y_axis[x]*128)+x_axis[x])
```

```python
#Adds new coloumn into dataframe
data1['Polarity']=polarity_number
```

```python
#Drops unwanted values/ Negative Polarity
data1.drop(data1.loc[data1['On/Off']==-1].index, inplace=True)
```
