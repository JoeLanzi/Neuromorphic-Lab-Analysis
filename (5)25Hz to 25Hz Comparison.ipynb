```python
import pandas as pd
import numpy as np
from plotly import __version__
import cufflinks as cf
from plotly.offline import download_plotlyjs,init_notebook_mode,plot,iplot
```

```python
init_notebook_mode(connected=True)
cf.go_offline()

%matplotlib inline
```

```python
data1 = pd.read_csv('25hz_test1.csv')

on_off1 = list(data1['On/Off'])
timestamp1 = list(data1[' Timestamp'])
```

```python
data2 = pd.read_csv('25hz_test2.csv')

on_off2 = list(data2['On/Off'])
timestamp2 = list(data2[' Timestamp'])
```

```python
first_time1=timestamp1[0]
for x in range(len(timestamp1)):
    timestamp1[x]=timestamp1[x]-first_time1
first_time2=timestamp2[0]
for x in range(len(timestamp2)):
    timestamp2[x]=timestamp2[x]-first_time2
```

```python
one_period = [i for i in timestamp1 if i < 28500] 
end = len(one_period)
end
```

```python
r1 = dict(zip(timestamp1[:end],on_off1[:end]))
r2 = dict(zip(timestamp2[:end],on_off2[:end]))
```

```python
df = pd.DataFrame(on_off1,index=timestamp1)
```

```python
df2 = pd.DataFrame(on_off2,index=timestamp2)
```

```python
result = pd.merge(df, df2, left_index=True, right_index=True)
```

```python
result[:500].iplot(kind='spread')
```

```python
answer = list((set(r1.items()) & set(r2.items())))
answer.sort()
len(answer)
```

```python
(len(answer)/len(one_period))*100
```

```python
percent = []

for item in range(100):
    first_time2=timestamp2[0]+item
    for x in range(len(timestamp2)):
        timestamp2[x]=timestamp2[x]-first_time2
        
    r1 = dict(zip(timestamp1[:100],on_off1[:100]))
    r2 = dict(zip(timestamp2[:100],on_off2[:100]))
    
    answer = list((set(r1.items()) & set(r2.items())))
    
    percent.append((len(answer)/end)*100)
```
